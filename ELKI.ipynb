```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./data_sg_filtered.csv",sep=' ')
temp = df[['factory','scan_spot']]
df = df.drop(columns=['factory','scan_spot'],axis=1)
df=(df-df.min())/(df.max()-df.min())
df = df.join(temp)

# group the data on Scan_spot value. 

spots = df.groupby('scan_spot')
dish = spots.get_group('Dish').drop(columns='scan_spot',axis=1)
flower = spots.get_group('Flower').drop(columns='scan_spot',axis=1)
white = spots.get_group('White').drop(columns='scan_spot',axis=1)

# Saving new features and labels

labels_d = dish['factory']
labels_f = flower['factory']
labels_w = white['factory']
dish = dish.drop(columns='factory',axis=1)
flower = flower.drop(columns='factory',axis=1)
white = white.drop(columns='factory',axis=1)
dish.to_csv("./results-sg-filtered/dish.csv", header=None, index=False)
flower.to_csv("./results-sg-filtered/flower.csv", header=None, index=False)
white.to_csv("./results-sg-filtered/white.csv", header=None, index=False)
labels_d.to_csv("./results-sg-filtered/labels_d.csv", header=None, index=False)
labels_f.to_csv("./results-sg-filtered/labels_f.csv", header=None, index=False)
labels_w.to_csv("./results-sg-filtered/labels_w.csv", header=None, index=False)
```
