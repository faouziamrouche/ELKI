```python
import pandas as pd
import numpy as np

df = pd.read_csv('data.csv',sep=';',encoding='latin-1')

df = df.rename(columns={"Immobilier": "Prix_immobilier"})
df = df.rename(columns={"id": "ID"})
df = df.drop(columns = ['Created Time'])
one_hot2 = pd.get_dummies(df['Nature operation'])
one_hot3 = pd.get_dummies(df['Type Co-emprunteur'])
one_hot4 = pd.get_dummies(df['Stage'])
one_hot5 = pd.get_dummies(df['Type de contrat'])
one_hot6 = pd.get_dummies(df['Profession'])
one_hot7 = pd.get_dummies(df['Domaine'])
one_hot8 = pd.get_dummies(df['civility'])
one_hot9 = pd.get_dummies(df['situation_familiale'])
one_hot10 = pd.get_dummies(df['regionName'])
# Drop old columns as they are now encoded
df = df.drop('Nature operation',axis = 1)
df = df.drop('Type Co-emprunteur',axis = 1)
df = df.drop('Stage',axis = 1)
df = df.drop('Type de contrat',axis = 1)
df = df.drop('Profession',axis = 1)
df = df.drop('Domaine',axis = 1)
df = df.drop('civility',axis = 1)
df = df.drop('situation_familiale',axis = 1)
df = df.drop('regionName',axis = 1)
# Join the encoded dataframes
df = df.join(one_hot2)
df = df.join(one_hot3)
df = df.join(one_hot4)
df = df.join(one_hot5)
df = df.join(one_hot6)
df = df.drop('Autre',axis = 1)
df = df.join(one_hot7)
df = df.join(one_hot8)
df = df.join(one_hot9)
df = df.join(one_hot10)
## Fill NaN with mean values
df.fillna(df.mean(), inplace=True)
## Display
df = df.drop_duplicates()
#df = df.sort_values(by=['ID'])
#df.to_csv("data-ready.csv")
df
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Shuttle_dataset_result.csv',sep=' ',comment='#',header=None)
#df = df.drop(10,axis = 1)
#df.to_csv('Shuttle_dataset_ready.csv')
df = df.drop_duplicates()
df
```

```python
train = df[:2000].sample(2000)
y_train = train['d_eligible']
X_train = train.drop(columns=['d_eligible','instruit'])
test = df[2000:].sample(450)
y_test = test['d_eligible']
X_test = test.drop(columns = ['d_eligible','instruit'])
```

```python
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
LR.predict(X_test)
print("LR : "+str(round(LR.score(X_test,y_test), 4)))

SVM = svm.SVC(decision_function_shape="ovo",gamma='auto').fit(X_train, y_train)
SVM.predict(X_test)
print("SVM : "+str(round(SVM.score(X_test, y_test), 4)))

RF = RandomForestClassifier(n_estimators=1000, max_depth=6, random_state=0).fit(X_train, y_train)
RF.predict(X_test)
print("RF : "+str(round(RF.score(X_test, y_test), 4)))

NN = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=1,max_iter=240).fit(X_train, y_train)
NN.predict(X_test)
print("NN on 240 epochs : "+str(round(NN.score(X_test, y_test), 4)))
```

```python
# result_instruit = xgb1.predict(test_X)
result = RF.predict_proba(X_test)
res = pd.DataFrame(result)
res.columns = ["Score_non_eligible","Score_eligible"]
res = res.sort_values(by=['Score_eligible'])
```
